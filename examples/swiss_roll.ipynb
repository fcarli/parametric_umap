```python
from parametric_umap import ParametricUMAP
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt
```

```python
n_samples = 1000
X, color = make_swiss_roll(n_samples=n_samples, noise=0, random_state=42)
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.viridis)

# Customize the plot
ax.set_title('Swiss Roll Dataset')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a colorbar
plt.colorbar(scatter)

plt.show()
```

```python
X_reduced = ParametricUMAP(device='cuda:3').fit_transform(X)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X_reduced[:,0], 
           X_reduced[:,1],
           c=color,
           cmap='viridis')
plt.colorbar(label='Color Value')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Reduced Coordinates with Color Mapping')
plt.show()
```
